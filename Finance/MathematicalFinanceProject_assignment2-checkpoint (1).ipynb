```python
import pandas as pd
import yfinance as yf
import datetime
import time
import requests
import io
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2021,2,1)
end = datetime.datetime(2022,2,1)
```

```python
tata_steel = pd.DataFrame() 
try:
    # download the stock price 
    stock = []
    stock = yf.download('TATASTEEL.NS',start=start, end=end, progress=False)
        
    # append the individual stock prices 
    if len(stock) == 0:
        None
    else:
        stock['Name']='TATASTEEL.NS'
        tata_steel = tata_steel.append(stock,sort=False)
except Exception:
    None
tata_steel
```

```python
aurobindo_pharma = pd.DataFrame()
try:
    # download the stock price 
    stock = []
    stock = yf.download('AUROPHARMA.NS',start=start, end=end, progress=False)
        
    # append the individual stock prices 
    if len(stock) == 0:
        None
    else:
        stock['Name']='AUROPHARMA.NS'
        aurobindo_pharma = aurobindo_pharma.append(stock,sort=False)
except Exception:
    None
aurobindo_pharma
```

```python
hdfc_bank = pd.DataFrame()
try:
    # download the stock price 
    stock = []
    stock = yf.download('HDFCBANK.NS',start=start, end=end, progress=False)
        
    # append the individual stock prices 
    if len(stock) == 0:
        None
    else:
        stock['Name']='HDFCBANK.NS'
        hdfc_bank = hdfc_bank.append(stock,sort=False)
except Exception:
    None
hdfc_bank
```

```python
stocks=pd.concat([tata_steel['Close'],aurobindo_pharma['Close'],hdfc_bank['Close']],axis=1)
stocks.columns=['tata_steel','aurobindo_pharma','hdfc_bank']
stocks
```

```python
returns=stocks/stocks.shift(1)
returns
```

```python
log_returns=np.log(returns)
log_returns
```

```python
number=10000
log_returns_mean=log_returns.mean()
weight=np.zeros((number,3))
expected_returns=np.zeros((number))
expected_volatility=np.zeros((number))
sharpe_ratio=np.zeros((number))
sigma=log_returns.cov()
```

```python
for k in range(number):
    w=np.array(np.random.random(3))
    wn=w/np.sum(w)
    weight[k,:]=wn
    
    expected_returns[k]=np.sum(log_returns_mean*w)
    
    expected_volatility[k]=np.sqrt(np.dot(w.T,np.dot(sigma,w)))
    
    sharpe_ratio[k]=expected_returns[k]/expected_volatility[k]
```

```python
max_index=sharpe_ratio.argmax()
print(max_index)
print(sharpe_ratio[max_index])
```

```python
plt.figure(figsize=(20,8))
plt.scatter(expected_volatility , expected_returns,c=sharpe_ratio)
plt.xlabel('expected volatility')
plt.ylabel('expected log return')
plt.title('Markowitz Frontier of my Portfolio')
plt.colorbar(label='Sharpe Ratio')
plt.scatter(expected_volatility[max_index], expected_returns[max_index],c='red')
plt.show()
```
